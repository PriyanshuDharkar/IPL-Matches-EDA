## Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

## Loading Data

```python
#loading matches data set
df_matches = pd.read_csv(r"C:\Users\ACER\OneDrive\Desktop\matches.csv")
```

## Exploratory Data Analysis

```python
#first five rows
df_matches.head()
```

## Handling missing values

```python
#missing values in matches data set
df_matches.isnull().sum()
```

```python
#drop umpire3 column
df_matches.drop("umpire3", axis = 1, inplace = True)
```

```python
#finding missing columns of city
df_matches[df_matches['city'].isnull()]
```

```python
#filling null values in city column
df_matches['city'] = df_matches['city'].fillna('Dubai')
```

## Handling missing values in umpire1, umpire2, player of the match columns

```python
#rows having missing values
df_matches[(df_matches['umpire1'].isnull()) | (df_matches['umpire2'].isnull()) | (df_matches['player_of_match'].isnull())]
```

```python
#deleting rows having missing values in umpire1, umpire2 and player_of_match
df_matches.dropna(subset = ['umpire1', 'umpire2', 'player_of_match'], inplace = True)
```

```python
df_matches.shape
```

## Number of teams participating each season

```python
#group by seasons and teams
df_matches_teams = df_matches.groupby('Season')['team1'].nunique()
```

```python
#plotting no. of teams participating each season
df_matches_teams.plot( kind = 'bar', figsize = (15,5), color = 'c')
plt.title("Number of teams participating each season", fontsize = 18, fontweight = 'bold')
plt.xlabel("Season", fontsize = 23)
plt.ylabel("Count of teams", fontsize = 23)
plt.show()
```

## Matches Played in Each Season


```python
#plotting no. of maches played each season
plt.figure(figsize = (18,6))
sns.countplot('Season', data = df_matches)
plt.title("Number of matches played in each IPL Season", fontsize = 20)
plt.xlabel("Season", fontsize = 15)
plt.ylabel("Matches", fontsize = 15)
plt.show()
```

## Number of Matches Won by Teams

```python
#plotting no. of matches won by teams
plt.figure(figsize = (18,6))
sns.countplot(x = 'winner', data = df_matches, palette = 'cool')
plt.title("Number of matches won by teams", fontsize = 20)
plt.xticks(rotation = 50)
plt.xlabel("Teams", fontsize = 15)
plt.ylabel("No. of Wins", fontsize = 15)
plt.show()
```

## Venue which has hosted most no. of matches

```python
df_matches['venue'].value_counts().sort_values(ascending = True).tail(10).plot( kind = 'barh', figsize = (12,8), fontsize = 15, color = 'purple')
plt.title("Venue which has hosted most number of matches", fontsize = 18, fontweight = 'bold')
plt.ylabel("Venue", size = 25)
plt.xlabel("Frequency", size = 25)
```

## Winners of IPL matches across 11 seasons

```python
df_matches['winner'].value_counts().sort_values(ascending = True).tail().plot(kind = 'barh', figsize = (15,5), color = 'grey')
plt.title("Winners of IPL matches across 11 seasons", fontsize = 18, fontweight = 'bold')
plt.xlabel("Frequency", size = 25)
plt.ylabel("Teams", size = 25)
plt.xticks(size = 15)
plt.yticks(size = 15)
```

## Toss decision

```python
colors = ['#FFBF00', '#FA8072']
df_matches['toss_decision'].value_counts().sort_values().plot(kind = 'pie', fontsize = 14, colors = colors, figsize = (10,7), autopct='%3.1f%%')
plt.ylabel("Toss Decision")
plt.title("Decision taken by captains after winning tosses", size = 20)
```

## Players with most MOTM awards

```python
df_matches['player_of_match'].value_counts().head(10).plot(kind = 'bar', figsize = (12,8), fontsize = 15, color = 'orange')
plt.title("Top 10 players with most MOTM Awards", fontsize = 18, fontweight = 'bold')
plt.xlabel("Players", size = 25)
plt.ylabel("Frequency", size = 25)
```

## Number of toss won by individual teams

```python
plt.figure(figsize = (18,6))
sns.countplot('toss_winner', data = df_matches, palette = "crest")
plt.xticks(rotation = 50)
plt.title("Number of toss won by teams", fontsize = 20, fontweight = 'bold')
plt.xlabel("Teams", size = 15)
plt.ylabel("No. of Toss", size = 15)
```

## Individual Teams decision to bat or field

```python
plt.figure(figsize = (18,6))
sns.countplot('toss_winner', data = df_matches, hue = 'toss_decision')
plt.xticks(rotation = 50)
plt.title("Teams decision to bat or field", fontsize = 20, fontweight = 'bold')
plt.xlabel("Toss Winner", size = 15)
plt.ylabel("Count", size = 15)
```
